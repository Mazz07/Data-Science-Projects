# Label encoding & Ordinal encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder # LabelEncoder / OrdinalEncoder
```

```python
df = pd.read_csv(r"F:\Data Science Projects\Data Cleaning\train.csv")
df.head()
```

```python
df2 = df[["KitchenQual","BldgType"]]
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(df2["BldgType"])
```

```python
df2["BldgType_L_enc"] = le.fit_transform(df2["BldgType"])
df2
```

```python
order_label = {'EX' : 4, 'Gd' : 3,'TA' : 2,'Fa' : 1}
```

```python
df2['KitchenQual_ord_enc'] = df2["KitchenQual"].map(order_label)
```
