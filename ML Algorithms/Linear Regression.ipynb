```python
import pandas as pd
import seaborn as sns
```

```python
#tips_df = pd.read_csv(r"F:\Data Science Projects\Data Cleaning\tips.csv")
tips_df = sns.load_dataset("tips")
tips_df
```

```python
x = pd.set_option("display.max_columns",None)
y = pd.set_option("display.max_rows",None)
```

```python
tips_df
```

```python
dummies = pd.get_dummies(tips_df)
dummies
```

```python
dummies.shape
```

```python
x = dummies.drop("total_bill",axis = 1)
x
```

# split data

```python
y = dummies["total_bill"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state = 51)
print("x_train shape = ",x_train.shape)
print("x_test shape = ",x_test.shape)
print("y_train shape = ",y_train.shape)
print("y_test shape = ",y_test.shape)
y_train
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ssc = StandardScaler()
ssc.fit(x_train)
x_train = ssc.transform(x_train)
x_test = ssc.transform(x_test)
x_train
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

# Predict values and test

```python
x_test[0, : ]
```

```python
lr.predict([x_test[0, : ]])
```

```python
lr.predict(x_test)
```

```python
y_test
```

```python
lr.score(x_test,y_test)
```

# Ridge and Lasso Regression

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rd = Ridge()
rd.fit(x_train,y_train)
rd.score(x_test,y_test)
```

```python
ls = Lasso()
ls.fit(x_train,y_train)
ls.score(x_test,y_test)
```

```python
rd2 = Ridge(alpha = 2)
rd2.fit(x_train,y_train)
rd2.score(x_test,y_test)
```

```python
ls2 = Lasso(alpha = 2)
ls2.fit(x_train,y_train)
ls2.score(x_test,y_test)
```

```python
ls5 = Lasso(alpha = 5)
ls5.fit(x_train,y_train)
ls5.score(x_test,y_test)
```

```python
rd5 = Ridge(alpha = 5)
rd5.fit(x_train,y_train)
rd5.score(x_test,y_test)
```

# Model Evaluation

# Root Mean Squre Error

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("mse = ",mse)
print("rmse = ",rmse)
```

# Polynomial Linear Regression - ML Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 2)
poly_reg.fit(x_train)
x_train_poly = poly_reg.transform(x_train)
x_test_poly = poly_reg.transform(x_test)
```

```python
x_train_poly.shape,x_test_poly.shape
```

```python
lr = LinearRegression()
lr.fit(x_train_poly,y_train)
```

```python
lr.score(x_test_poly,y_test)
```

```python
lr.predict([x_test_poly[0,:]])
```

```python
y_pred = lr.predict(x_test_poly)
y_pred
```

```python
y_test
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("mse = ",mse)
print("rmse = ",rmse)
```
