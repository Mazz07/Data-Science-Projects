```python
import pandas as pd 
```

```python
df = pd.read_csv(r"F:\Data Science Projects\WorldCups.csv")
df
```

```python
df2 = df[["GoalsScored","QualifiedTeams","MatchesPlayed","Year"]]
df2
```

```python
x = df2.drop("QualifiedTeams",axis = 1)
x
```

```python
y = df2["QualifiedTeams"]
y
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test = train_test_split(x,y,test_size = 0.5,random_state = 52)
print("shape of x_train",x_train.shape)
print("shape of x_test",x_test.shape)
print("shape of y_train",y_train.shape)
print("shape of y_test",y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ssc = StandardScaler()
ssc.fit(x_train)
x_train = ssc.transform(x_train)
x_test = ssc.transform(x_test)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```
