# Histogram plot


```python
import matplotlib.pyplot as plt
import numpy as np
import random
ml_students_age = np.random.randint(18,50,(100))
py_students_age = np.random.randint(15,40,(100))
from matplotlib import style
bins = [15,20,25,30,35,40,45,50]
plt.figure(figsize=(16,7))
plt.hist([ml_students_age,py_students_age],bins, rwidth=0.8,color = ["r","m"],label = ["ML_students_age","py_students_age"])
#plt.hist(py_students_age,bins, rwidth=0.8,color = "m",label = "py_students_age")
style.use("ggplot")
plt.title("Students_age")
plt.xlabel("age catagory of students")
plt.ylabel("No.of students")
plt.grid(color='w', linestyle='-.', linewidth=1)
plt.legend()
plt.show()
```

# Bar chart plot

```python
import matplotlib.pyplot as plt
import numpy as np 
from matplotlib import style
classes = ['python','ML','AI','SQL','R']
class_1_students = [30,12,23,27,8]
class_2_students = [20,11,45,16,38]
class_3_students = [40,22,19,21,50]
classes_index = np.arange(len(classes))
width = 0.2
plt.figure(figsize=(16,9))
plt.bar(classes_index,class_1_students,color="r",width=0.2,align="center",edgecolor="k",linewidth="1",label='class_1_students' )
plt.bar(classes_index + width,class_2_students,color="m",width=0.2,align="center",edgecolor="k",
        linewidth="1",label='class_2_students' )
plt.bar(classes_index + width + width,class_3_students,color="g",width=0.2,align="center",edgecolor="k",linewidth="1",
        label='class_3_students' )
plt.title('class',fontsize=30)
plt.xticks(classes_index + width,classes,rotation = 20)
plt.xlabel('classes',fontsize=20)
plt.ylabel('class_students',fontsize=20)
style.use("ggplot")
plt.legend(loc=1)
plt.show()
```

# pie plot

```python
import matplotlib.pyplot as plt
```

```python
players = ['Ronaldo','Messi','Neymar','Mbappe','Benzema']
goals = [850,800,745,648,545]
explode = [0.08,0.01,0,0,0]
colors = ['r','g','b','c','m']
textprops = {"fontsize" : 15}
wedgeprops = {"linewidth":2, "width":1, "edgecolor":"k"}
plt.title('Goals Scored')
plt.pie(goals, labels=players,explode = explode,colors = colors,autopct="%0.2f%%",wedgeprops=wedgeprops,pctdistance=0.7,
        counterclock=True,shadow=False,textprops = textprops,radius=1,startangle=90)
plt.show()
```
