```python
import seaborn as sns
import pandas as pd
import scipy.stats as norm
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import style
```

```python
goals = [850,800,745,648,545,624,405,365,581]
networth = [770,650,600,590,530,580,550,540,588]
players_data = ({"goals":goals,"networth":networth})
sns.lineplot(x="goals", y="networth", data = players_data)
plt.show()
```

```python
tips_data=sns.load_dataset("tips")
tips_data
```

```python
sns.lineplot(x="size",y="tip",data = tips_data)
plt.show()
```

```python
sns.lineplot(x="size",y="tip",data = tips_data,hue="sex",palette="hot",style="sex",markers=["o",">"],)
plt.show()
```

```python
sns.lineplot(x="size",y="tip",data = tips_data,hue="day",palette="hot",style="day",markers=["o",">","<","^"],)
plt.show()
```

```python
sns.distplot(tips_data["total_bill"],bins=10,hist=True,kde=True,rug=True,color="r",label="Total Bill")
plt.show()
```

```python
sns.set()
sns.distplot(tips_data["total_bill"],bins=10,hist=True,kde=True,rug=True,color="r",label="Total Bill",)
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55,60]
sns.distplot(tips_data["total_bill"],bins=bins,hist=True,kde=True,color="r")
plt.show()
```

```python
sns.distplot(tips_data["total_bill"],hist_kws={"color":"r","edgecolor":"y","linewidth":2,"linestyle":"--","alpha":1},
            kde_kws={"color":"k","linewidth":2,"linestyle":"--","alpha":0.8},rug = True,
               rug_kws={"color":"g","linewidth":2,"linestyle":"--","alpha":0.4},
                fit_kws={"color":"k","linewidth":4,"linestyle":"--","alpha":1})
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55,60]
sns.distplot(tips_data["total_bill"],bins=bins,label="total_bill")
sns.distplot(tips_data["tip"] ,bins=bins,label="tip")
sns.distplot(tips_data["size"],bins=bins,label="size")
plt.legend()
plt.show()
```
